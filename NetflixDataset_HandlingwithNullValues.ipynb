**Fillna** fonksiyonunu ile null değerden **bir önceki** değeri null değerin yerine alabiliriz.




```python
df4= df.fillna(method="pad")
df4

#pad ile bir önceki (sütunda) datayı null değerin yerine alabiliyoruz. Anak aşağıdaki örnekte de görüldüğü gibi 0. satırda NaN
#değer olması bu dataset için uygun bir yöntem olmadığını anlayabiliyoruz.
```

```python
df4.isnull().sum().sum()

#0.satırdaki NaN değer :) (Some NaN value problems asdfghjk)
```

**Fillna** methodu ile NaN değeri olan sütunlardaki NaN değerlerin yerine sütunun ortalama değeri yazılıyor

Dataset içerisinde null değerlerin bulunduğu sütunlar genel olarak string datalarda. Yeni bir değişkene datasetin bir kısmını
atayarak bu datasete random null değerler atayabiliriz . Bu sayede null değerlerin bulunduğu sütunlar sayısal veri içeriyorsa
nasıl handle edebileceğimizi de anlamış oluruz.


```python
netflix_copy=df.copy()
```

```python
netflix_copy
```

```python
import numpy as np

netflix_copy.loc[np.random.randint(1, 7787 , 20),"duration"] = None

#duration sütununa random olarak NaN değerler atıyoruz.
#1 den 7787 ye kadar 20 tane random nan değer ataması yapıyoruz.
```

```python
netflix_copy["duration"].isnull().sum()
```

```python
df["duration"].isnull().sum()
```

```python
netflix_copy["duration"].isnull().index
```

**Artık sayısal bir sütunda NaN değerler var ve böyle bir durumda nasıl NaN değerleri handle edebiliriz ?**

```python
netflix_copy_index=netflix_copy.duration.sort_values().tail(10).index
netflix_copy_index
```

```python
netflix_copy2 = netflix_copy.fillna(value = netflix_copy["duration"].mean())
netflix_copy2

#Fillna methodu ile NaN değeri olan sütunlardaki NaN değerlerin yerine sütunun ortalama değeri yazılıyor
```

```python
netflix_copy.loc[netflix_copy_index]
```

```python
netflix_copy2.loc[netflix_copy_index]
```

```python
netflix_copy3=netflix_copy.fillna(value=df["duration"].max())
netflix_copy3
```

```python
netflix_copy3.loc[netflix_copy_index]
```

```python
netflix_copy4=df.fillna(value=df["duration"].min())
netflix_copy4
```

```python
netflix_copy4.loc[netflix_copy_index]
```
