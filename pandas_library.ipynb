```python
films=np.array(["A Beautiful Mind"," American Beauty","Godfather","Big Fish","Blade Runner","Blue Velvet",
                "Clockwork Orange","Dog Day Afternoon","Full Metal Jacket","Lord of The Rings"])
genres=np.array(["Drama","Romance","Drama","Romance",",Science Fiction","Romance","Dram","Dram","Action","Fantastic"])

lead_studios=np.array(["Warner Bros.","Warner Bros.","Sony","Warner Bros.","Warner Bros.","Independent",
                      "Independent","Sony","Independent","Warner Bros."])

gross=np.array([100000,200000,1000000,150000,1000000,50000,20000,500000,10000,10000000])


```

```python
dictionary={"Film":films,
           "Genre":genres,
           "Lead Studio":lead_studios,
           "Worldwide Gross":gross,
           
           }
```

```python
current_year=pd.DataFrame(dictionary)
current_year.head()
```

```python
current_year["Year"]=2023
current_year
```
